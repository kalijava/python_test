```python

import random
import string


string1=input("entrer une chaine de caractere de votre choix...")
random1 = ''.join([random.choice(string.ascii_letters
			+ string.digits) for n in range(len(string1))])
i=1
while (string1 != random1):
    print(random1)
    i+=1
    random1 = ''.join([random.choice(string.ascii_letters
			+ string.digits) for n in range(len(string1))])
print(random1)
print("cible atteint apres ",i,"iterations")
    




```
